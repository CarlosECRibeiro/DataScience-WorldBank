## Conclusão do ensino primário Feminio

Aqui foi analizada a Taxa de conclusão do ensino primário feminino ao longo dos anos e apresentados na forma de gráfico das segunites maneiras:

1. Linha
2. Barra
3. Histograma
4. Pizza

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fornecidos
df = pd.read_csv("./data.csv").head(20)

# Gráfico de Linha
plt.figure(figsize=(30, 12))
for i in range(len(df)):
    plt.plot(df.columns[4:], df.iloc[i, 4:], marker='o', label=df.iloc[i, 0])
plt.title('Taxa de conclusão do ensino primário feminino ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('Taxa de conclusão (%)')
plt.legend()
plt.grid(True)


# Gráfico de Barra
plt.figure(figsize=(30, 12))
year = '2020'  # Escolhendo o ano para análise
plt.bar(df['Country Name'], df[year])
plt.title(f'Taxa de conclusão do ensino primário feminino em {year}')
plt.xlabel('País ou Região')
plt.ylabel('Taxa de conclusão (%)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Histograma
plt.figure(figsize=(30, 12))
year = '2020'  # Escolhendo o ano para análise
plt.hist(df[year].dropna(), bins=10, edgecolor='black')
plt.title(f'Distribuição da taxa de conclusão do ensino primário feminino em {year}')
plt.xlabel('Taxa de conclusão (%)')
plt.ylabel('Frequência')
plt.grid(axis='y')
plt.tight_layout()
plt.show()


# Aqui filtramos os dados para incluir apenas os países com valores válidos para o ano escolhido
year = '2020'  # Aqui escolhemos o ano para análise
data_valid = df.dropna(subset=[year])

# Gráfico de Pizza
plt.figure(figsize=(10, 6))
plt.pie(data_valid[year], labels=data_valid['Country Name'], autopct='%1.1f%%', startangle=140)
plt.title(f'Contribuição percentual para a taxa de conclusão do ensino primário feminino em {year}')
plt.axis('equal')
plt.tight_layout()
plt.show()

```
